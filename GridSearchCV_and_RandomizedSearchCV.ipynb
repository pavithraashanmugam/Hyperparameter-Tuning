HYPERPARAMETER TUNING TYPES:
1. GridSearchCV
2. RandomizedSearchCV

Importing dependencies

```python
import numpy as np
import pandas as pd
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

Loading the dataset

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
df = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
df.head()
```

Adding the target col to the dataframe

```python
df['label']=breast_cancer_dataset.target
```

```python
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df['label'].value_counts()
```

Seperating the feture and the target

```python
X = df.drop(columns='label', axis=1)
Y = df['label']
```

```python
print(X)
print(Y)
```

Converting datas to array

```python
X = np.asarray(X)
Y = np.asarray(Y)
```

```python
print(X)
print(Y)
```

# GRID_SEARCH_CV



It is used to determine the best parameter for the model

```python
# loading SVC model
model = SVC()
```

```python
# hyperparameter
parameters = {
              'kernel':['linear', 'rbf', 'poly', 'sigmoid'],
              'C' : [1, 5, 10, 20]
              }
```

```python
# grid search
classifier = GridSearchCV(model, parameters, cv=5)
```

```python
# fitting the data to the model
classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
# best parameter
best_params = classifier.best_params_
print(best_params)
```

```python
# determining highest accuracy
highest_accuracy = classifier.best_score_
print(highest_accuracy)
```

```python
# loading the results to pd dataframe
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
grid_search_result = result[['param_C','param_kernel','mean_test_score' ]]
```

```python
grid_search_result
```

Highest accuracy = 95.2 %   
Best parameter:   
    

*   C = 10
*   Kernel = linear




# RANDOMIZED_SEARCH_CV

Randomly finds best parameters

```python
model = SVC()
```

```python
# hyperparameter
parameters = {
              'kernel':['linear', 'rbf', 'poly', 'sigmoid'],
              'C' : [1, 5, 10, 20]
              }
```

```python
# randomized search
classifier = RandomizedSearchCV(model, parameters, cv=5)
```

```python
# fitting the data to the model
classifier.fit(X, Y)
```

```python
classifier.cv_results_
```

```python
classifier.best_params_
```

```python
classifier.best_score_
```

```python
# loading the result to the pd dataframe
result = pd.DataFrame(classifier.cv_results_)
```

```python
result.head()
```

```python
randomized_search_result = result[['param_C','param_kernel','mean_test_score']]
```

```python
randomized_search_result
```
